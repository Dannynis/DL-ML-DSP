```python
from __future__ import division

import scipy as sp
from scipy.optimize import minimize
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
```

# 1 Logistic regression

```python
data = sp.loadtxt('ex2/ex2data1.txt', delimiter=',')
```

```python
exam_1 = data[:, 0]
exam_2 = data[:, 1]

X = sp.hstack((sp.ones((len(exam_1), 1)), exam_1.reshape(-1,1), exam_2.reshape(-1,1))).T
y = data[:, 2]
m = len(y) 
```

## 1.1 Visualizing the data

```python
plt.figure()
plt.plot(exam_1[y == 0], exam_2[y == 0], 'yo', ms=8, label='Not admitted')
plt.plot(exam_1[y == 1], exam_2[y == 1], 'k+', ms=8, label='Admitted')
plt.xlabel('Exam 1 score')
plt.ylabel('Exam 2 score')
plt.xlim(28, 119)
plt.ylim(28, 119)
plt.legend(numpoints = 1, loc = 1)
plt.savefig('ex2_ipynb_img/fig1.jpg')
plt.show()
```

## 1.2 Implementation

```python
sigmoid = lambda x: 1 / (1 + sp.exp(-x))

def cost_function(theta, X, y):
    cost = (1 / m) * (- y.dot(sp.log(sigmoid(theta.dot(X)))) - (1 - y).dot(sp.log(1 - sigmoid(theta.dot(X)))))
    grad = (1 / m) * (sigmoid(theta.dot(X)) - y).dot(X.T)
    return cost, grad
```

```python
theta_0 = sp.zeros(3)
cost_function(theta_0, X, y)
```

```python
xks = []
def callback(xk):
    xks.append(xk);
    
res = minimize(cost_function, theta_0, method = 'TNC', args=(X, y), jac=True, callback=callback) 
```

```python
theta = res['x']
print theta
print cost(theta)
```

```python
J_history = [cost_function(item, X, y)[0] for item in xks]

plt.figure()
plt.plot(J_history, 'ko')
plt.xlabel('Steps')
plt.ylabel('Cost')
plt.xlim(-res['nit'] * 0.05, res['nit'] * 1.05)
plt.ylim(0, max(J_history) * 1.05)
plt.savefig('ex2_ipynb_img/cost_vs_steps.jpg')
plt.show()
```

```python
def prediction(exam_1, exam_2, theta):
    return sigmoid(theta[0] + exam_1 * theta[1] + exam_2 * theta[2])
```

```python
prediction(45, 85, theta)
```

```python
def accuracy(exam_1, exam_2, y, theta):
    return sp.sum(sp.around(prediction(exam_1, exam_2, theta)) == y) / m
```

```python
accuracy(exam_1, exam_2, y, theta)
```
